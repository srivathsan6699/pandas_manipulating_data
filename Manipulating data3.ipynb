
## Shuffle , Reset , function apply

```python
import pandas as pd
```

```python
a = pd.read_csv("car-sales-missing-data.csv")
a
```

## To shuffle the dataframe

```python
a_shuffledd = a.sample(frac=1) # if i give frac = 0.5 then it will shuffle half the dataframe
a_shuffledd #This will shuffle the data frame and prints result 
```

## Resetting the shuffled index

```python
a_shuffledd.reset_index(drop = True, inplace = True) # this will reset the shuffled index 
a_shuffledd
```

```python
## Lambda is a keyword in python which means anonymous fuction
a_shuffledd["Odometer"] = a_shuffledd["Odometer"].apply(lambda x:x/1.6) #In this lambda is used to reassign x(ie odometer) to odometer/1.6
a
```
